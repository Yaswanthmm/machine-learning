```python
# a. Using NumPy create random vector of size 15 having only Integers in the range 1-20.
import numpy as np
from numpy import random
rand_int = np.random.randint(low=1, high=21, size=15)
print(rand_int)
```

```python
# 1.Reshape the array to 3 by 5
arr = rand_int.reshape(3, 5)
print (arr)
# 2.Print array shape.
print('\n array shape: ',arr.shape)
#3.Replace the max in each row by 0
max= np.argmax(arr, axis=1, out=None)
x=0
for i in max:
    arr[x][i] = 0
    x+=1
arr
```

```python
# Create a 2-dimensional array of size 4 x 3 (composed of 4-byte integer elements), also print the shape, type and data type of the array
from numpy import random

arr2 = np.array(random.randint(1,21,size=(4,3)), np.int32)
print("array shape:", arr2.shape)
print("array type:", type(arr2))
print("Data type:", arr2.dtype)
```

```python
# b. Write a program to compute the eigenvalues and right eigenvectors of a given square array given below:
arr3 = np.array([[3, -2], [1, 0]])
eigenvalues, eigenvectors = np.linalg.eig(arr3)
print("\n Eigenvalues: ", eigenvalues)
print("\n Eigenvectors: ", eigenvectors)
```

```python
# c. Compute the sum of the diagonal element of a given array.
arr4 = np.array([[0, 1, 2], [3, 4, 5]])
sum_diagonal = np.trace(arr4)
sum_diagonal
```

```python
# d. Write a NumPy program to create a new shape to an array without changing its data.
arr5 = np.array([[1, 2], [3, 4], [5, 6]])
print(arr5)
new_shape1 = arr5.reshape(2,3)
print("\n new shape")
print(new_shape1)
arr6 = np.array([[1, 2, 3], [4, 5, 6]])
new_shape2 = arr6.reshape(3,2)
print("\n new shape")
print(new_shape2)
```
